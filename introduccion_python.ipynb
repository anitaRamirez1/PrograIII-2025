Hacer un ejercicio en python usando la funcion filter y lambda que de una seria de notas introducidas por el usuario, establezca cuantos pasan la materia (>6) y cuantos se quedan (<6).



```python
#reduce
from functools import reduce

numeros = [5,4,9,6,8,2,3]
suma = reduce(lambda acum, num: num+acum, numeros)

sumaPares = reduce(lambda acum, num: num+acum, filter(lambda num: num%2==0, numeros))
sumaPar = reduce(lambda acum, num: acum + num if num%2==0 else acum, numeros,0)

print("La suma es: ", suma)
print("La suma de los pares es: ", sumaPares)
print("Suma Pares: ", sumaPar)
```

```python
#Hecer un programa que lea una serie de numero introeducidos por el usuario y determine la media artimetica.
from math import sqrt
class estadistica:
  def media(self, serie):
    suma = 0
    for num in serie:
      suma += int(num)
    return suma/len(serie)

  def tipica(self, serie):
    media = self.media(serie)
    suma = 0
    for num in serie:
      suma += (int(num)-media)**2
    return sqrt( suma/len(serie) )
```

```python
serie = input("Serie ejem: 1,5,2,3: ").split(",")
objEstadistica = estadistica()
print("La media es: ", objEstadistica.media(serie), "la desvicacion tipica es: ", round(objEstadistica.tipica(serie), 2) )
```

```python
def calcular_impuesto_actividades_economicas(monto_actividad):
    """
    Calcula el IMPUESTO A LAS ACTIVIDADES ECONÓMICAS de acuerdo con la tabla dada.
    """
    tabla_impuestos = [
        (0.01, 500, 1.5, 0),
        (500.01, 1000, 1.5, 3),
        (1000.01, 2000, 3, 3),
        (2000.01, 3000, 6, 3),
        (3000.01, 6000, 9, 2),
        (8000.01, 18000, 15, 2),
        (18000.01, 30000, 39, 2),
        (30000.01, 60000, 63, 1),
        (60000.01, 100000, 93, 0.8),
        (100000.01, 200000, 125, 0.7),
        (200000.01, 300000, 195, 0.6),
        (300000.01, 400000, 255, 0.45),
        (400000.01, 500000, 300, 0.4),
        (500000.01, 1000000, 340, 0.30),
        (1000000.01, 99999999, 490, 0.18)
    ]

    for desde, hasta, precio, adicional in tabla_impuestos:
        if desde <= monto_actividad <= hasta:
            if adicional == 0:
                return precio
            else:
                impuesto = ((monto_actividad - desde) / 1000 * adicional) + precio
                return round(impuesto, 2)
    return "Monto fuera de los rangos definidos."

def convertir_area(cantidad, origen, destino):
    """
    Convierte unidades de área basadas en las medidas de El Salvador.
    """
    unidades = {
        "pie cuadrado": 0.092903,
        "vara cuadrada": 0.6987,
        "yarda cuadrada": 0.836127,
        "metro cuadrado": 1.0,
        "tarea": 437.5, # Valor corregido: 1/16 de una manzana.
        "manzana": 7000.0, # 7,000 m2
        "hectárea": 10000.0
    }

    origen_lower = origen.lower()
    destino_lower = destino.lower()

    if origen_lower not in unidades or destino_lower not in unidades:
        return None

    cantidad_en_m2 = cantidad * unidades[origen_lower]
    resultado = cantidad_en_m2 / unidades[destino_lower]
    return resultado

def simular_cajero(monto_a_retirar):
    """
    Simula la dispensación de billetes y monedas.
    """
    try:
        cantidad = float(monto_a_retirar)
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número.")
        return

    if cantidad <= 0:
        print("El monto a retirar debe ser un número positivo.")
        return

    print(f"\nDesglose para ${cantidad:.2f}:")
    denominaciones_billetes = [100, 50, 20, 10, 5, 1]
    denominaciones_monedas = [0.25, 0.10, 0.05, 0.01]
    cantidad_restante = cantidad

    for billete in denominaciones_billetes:
        if cantidad_restante >= billete:
            cantidad_billetes = int(cantidad_restante // billete)
            print(f"{cantidad_billetes} billete(s) de a ${billete}")
            cantidad_restante %= billete

    cantidad_restante_centavos = round(cantidad_restante * 100)
    for moneda_centavos in [int(m * 100) for m in denominaciones_monedas]:
        if cantidad_restante_centavos >= moneda_centavos:
            cantidad_monedas = int(cantidad_restante_centavos // moneda_centavos)
            valor_moneda_dolar = moneda_centavos / 100
            print(f"{cantidad_monedas} moneda(s) de a ${valor_moneda_dolar:.2f} ctvs.")
            cantidad_restante_centavos %= moneda_centavos

def main():
    while True:
        print("\n--- Parcial I INGENIERIA EDU-ANA ---")
        print("1. Cálculo de Impuesto ")
        print("2. Conversor de Área ")
        print("3. Cajero Automático")
        print("4. Exit")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            try:
                monto_input = float(input("Ingrese el monto de la actividad económica: $"))
                impuesto = calcular_impuesto_actividades_economicas(monto_input)
                if isinstance(impuesto, str):
                    print(f"Error: {impuesto}")
                else:
                    print(f"El valor a pagar es: ${impuesto:.2f}")
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número.")
        elif opcion == "2":
            print("Unidades disponibles: Pie Cuadrado, Vara Cuadrada, Yarda Cuadrada, Metro Cuadrado, Tarea, Manzana, Hectárea.")
            try:
                cantidad_area = float(input("Ingrese la cantidad a convertir: "))
                unidad_origen = input("Ingrese la unidad de origen: ")
                unidad_destino = input("Ingrese la unidad de destino: ")
                resultado_area = convertir_area(cantidad_area, unidad_origen, unidad_destino)
                if resultado_area is None:
                    print("Una o ambas unidades de área no son válidas. Por favor, verifique.")
                else:
                    print(f"{cantidad_area} {unidad_origen.capitalize()} = {resultado_area:.4f} {unidad_destino.capitalize()}")
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número para la cantidad.")
        elif opcion == "3":
            try:
                monto_retirar = float(input("Ingrese la cantidad a retirar: $"))
                simular_cajero(monto_retirar)
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número.")
        elif opcion == "4":
            print("Gracias por usar el programa.")
            break
        else:
            print("Opción inválida. Intente nuevamente.")

if __name__ == "__main__":
    main()
```
