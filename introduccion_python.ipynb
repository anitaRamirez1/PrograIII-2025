```python
#calcular el sueldo de un empleado segun el numero de horas trabajadas yel valor por horas


def sueldo(numHrs, valorXhrs):
  sueldo = numHrs*valorXhrs * 30

  return sueldo
```

```python
numHrs = float(input(" Hrs: "))
valorXhrs = float(input(" valor Hrs: "))
sueldo = sueldo(numHrs, valorXhrs)
print("sueldo: ", sueldo)
```
