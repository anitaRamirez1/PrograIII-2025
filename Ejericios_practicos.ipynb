```python
tasas = {
    "USD": 1.0,
    "SVC": 8.74,      # colón salvadoreño
    "EUR": 0.85,      # euro
    "GBP": 0.74,      # libra esterlina
    "JPY": 148.85,     # yen japonés
    "MXN": 18.74,      # peso mexicano
    "CAD": 1.38,      # dólar canadiense
    "CNY": 7.19,       # yuan chino
    "PEN": 3.56,       # sol peruano
    "CRC": 505.13      # colón costarricense
}

def convertir_moneda(cantidad, origen, destino):
    origen = origen.upper()
    destino = destino.upper()
    if origen not in tasas or destino not in tasas:
        return None, "Una o ambas monedas no son válidas."
    # Convertir origen -> USD -> destino
    monto_usd = cantidad / tasas[origen]
    convertido = monto_usd * tasas[destino]
    return convertido, None

def conversor_monedas():
    print("=== Conversor de Monedas ) ===")
    print("Monedas disponibles:", ", ".join(tasas.keys()))

    origen = input("Moneda de ORIGEN: ").strip().upper()
    destino = input("Moneda de DESTINO: ").strip().upper()
    try:
        cantidad = float(input("Cantidad a convertir: "))
    except ValueError:
        print("Cantidad inválida. Debe ser un número.")
        return

    resultado, error = convertir_moneda(cantidad, origen, destino)
    if error:
        print(error)
    else:
        print(f"{cantidad:.2f} {origen} = {resultado:.2f} {destino}")


#if __name__ == "__main__":
  #  main()
```

```python


def conversor_longitud():
    unidades = {
        "metro": 1,
        "kilometro": 1000,
        "centimetro": 0.01,
        "milimetro": 0.001,
        "micrometro": 1e-6,
        "nanometro": 1e-9,
        "milla": 1609.34,
        "yarda": 0.9144,
        "pie": 0.3048,
        "pulgada": 0.0254
    }

    print("\n--- Conversor de Longitud ---")
    print("Unidades disponibles:")
    for u in unidades.keys():
        print("-", u)

    # Pedir datos al usuario
    origen = input("\nIngrese la unidad de origen: ").lower()
    destino = input("Ingrese la unidad de destino: ").lower()
    cantidad = float(input("Ingrese la cantidad a convertir: "))

    # Validar unidades
    if origen not in unidades or destino not in unidades:
        print(" Unidad no válida, intente otra vez.")
        return

    # Conversión: primero a metros, luego a destino
    en_metros = cantidad * unidades[origen]
    resultado = en_metros / unidades[destino]

    print(f"\n {cantidad} {origen} = {resultado:.3f} {destino}")

# Ejecutar
conversor_longitud()
```

```python
unidades = {
    "KG": ("Kilogramos", 1.0),
    "G": ("Gramos", 1000.0),
    "MG": ("Miligramos", 1_000_000.0),
    "T": ("Toneladas", 0.001),
    "LB": ("Libras", 2.20462),
    "OZ": ("Onzas", 35.274),
    "ST": ("Stone", 0.157473),
    "Q": ("Quintales", 0.01),
    "MCG": ("Microgramos", 1_000_000_000.0),
    "DAG": ("Decagramos", 100.0)
}

def convertir_masa(cantidad, origen, destino):
    origen = origen.upper()
    destino = destino.upper()
    if origen not in unidades or destino not in unidades:
        return None, "Una o ambas unidades no son válidas."

    # Convertir origen -> kilogramos -> destino
    valor_kg = cantidad / unidades[origen][1]
    convertido = valor_kg * unidades[destino][1]
    return convertido, None

def conversor_masa():
    print("=== Conversor de Masa ===")
    print("Unidades disponibles:")
    for codigo, (nombre, _) in unidades.items():
        print(f"{codigo}: {nombre}")

    origen = input("Unidad de ORIGEN (código): ").strip().upper()
    destino = input("Unidad de DESTINO (código): ").strip().upper()

    try:
        cantidad = float(input("Cantidad a convertir: "))
    except ValueError:
        print("Cantidad inválida. Debe ser un número.")
        return

    resultado, error = convertir_masa(cantidad, origen, destino)
    if error:
        print(error)
    else:
        print(f"{cantidad} {unidades[origen][0]} ({origen}) = {resultado} {unidades[destino][0]} ({destino})")

conversor_masa()
```

```python
def conversor_almacenamiento():

    unidades = {
        "bit": 1 / 8,                     # 1 bit = 1/8 bytes
        "byte": 1,                        # 1 byte = 1 byte
        "kilobyte": 1_000,                # 1 KB = 1,000 bytes
        "megabyte": 1_000_000,              # 1 MB = 1,000,000 bytes
        "gigabyte": 1_000_000_000,            # 1 GB = 1,000,000,000 bytes
        "terabyte": 1_000_000_000_000,          # 1 TB = 1,000,000,000,000 bytes
        "petabyte": 1_000_000_000_000_000,
        "exabyte": 1_000_000_000_000_000_000,
        "zettabyte": 1_000_000_000_000_000_000_000,
        "yottabyte": 1_000_000_000_000_000_000_000_000
    }

    print("\n--- Conversor de Almacenamiento ---")
    print("Unidades disponibles:")
    for u in unidades.keys():
        print("-", u)

    # Pedir datos al usuario
    origen = input("\nIngrese la unidad de origen: ").lower()
    destino = input("Ingrese la unidad de destino: ").lower()

    try:
        cantidad = float(input("Ingrese la cantidad a convertir: "))
    except ValueError:
        print("Cantidad inválida. Debe ser un número.")
        return

    # Validar unidades
    if origen not in unidades or destino not in unidades:
        print("Unidad no válida, intente otra vez.")
        return

    # Conversión: primero a bytes, luego a unidad de destino
    en_bytes = cantidad * unidades[origen]
    resultado = en_bytes / unidades[destino]

    print(f"\n{cantidad} {origen} = {resultado:.6f} {destino}")

# Ejecutar
conversor_almacenamiento()
```

```python
def conversor_area_el_salvador():
    # Factores de conversión actualizados y verificados para El Salvador
    unidades = {
        "milimetro cuadrado": 0.000001,
        "centimetro cuadrado": 0.0001,
        "metro cuadrado": 1,
        "hectarea": 10000,
        "acre": 4046.8564224,
        "manzana": 6988.96,             # 1 manzana = 6,988.96 m² (7000 varas²)
        "tarea": 628.864,               # 1 tarea = 628.864 m² (625 varas²)
        "vara cuadrada": 0.698896,
        "caballeria": 447900,           # CORRECCIÓN: 1 caballería = 44.79 ha = 447,900 m²
        "pie cuadrado": 0.09290304,
        "hectometro cuadrado": 10000
    }

    print("\n=== CONVERSOR DE ÁREA - EL SALVADOR (Verificado) ===")
    print("\nUnidades disponibles:")
    print("----------------------------------------")
    for idx, unidad in enumerate(unidades.keys(), 1):
        print(f"{idx:2d}. {unidad}")
    print("----------------------------------------")
    #print("NOTA: 1 caballería = 64 manzanas = 44.79 ha (estándar El Salvador)")

    try:
        origen_idx = int(input("\nSeleccione número de unidad de origen: "))
        destino_idx = int(input("Seleccione número de unidad de destino: "))

        if not (1 <= origen_idx <= len(unidades)) or not (1 <= destino_idx <= len(unidades)):
            print("Error: Debe seleccionar números entre 1 y", len(unidades))
            return

        origen = list(unidades.keys())[origen_idx-1]
        destino = list(unidades.keys())[destino_idx-1]

        cantidad = float(input(f"Ingrese la cantidad en {origen}: "))
        if cantidad < 0:
            print("Error: La cantidad no puede ser negativa")
            return

    except ValueError:
        print("Error: Debe ingresar valores numéricos válidos")
        return

    # Conversión exacta
    en_metros = cantidad * unidades[origen]
    resultado = en_metros / unidades[destino]

    print(f"\n► RESULTADO: {cantidad} {origen} = {resultado:,.4f} {destino}")
   # print(f"Equivalencia exacta: 1 caballería = 44.79 hectáreas (estándar El Salvador)")

if __name__ == "__main__":
    conversor_area_el_salvador()
```

```python
def mostrar_menu():

    print("\n--- MENÚ DE CONVERSIÓN ---")
    print("1. Monedas")
    print("2. Longitud")
    print("3. Masa")
    print("4. Almacenamiento")
    print("5. Tiempo")
    print("6. Volumen")
    print("7. Área")
    print("8. Salir")
    print("--------------------------")

# Bucle principal del programa
while True:
    mostrar_menu()
    opcion = input("Por favor, selecciona una opción (1-8): ")

    if opcion == '1':
       conversor_monedas()
    elif opcion == '2':
        conversor_longitud()
    elif opcion == '3':
        conversor_masa()
    elif opcion == '4':
        conversor_almacenamiento()
    elif opcion == '5':
       conversor_tiempo()
    elif opcion == '6':
       conversor_volumen()
    elif opcion == '7':
        conversor_area_el_salvador()
    elif opcion == '8':
        print("Saliendo del programa. ¡Hasta pronto!")
        break  # Rompe el bucle para salir del programa
    else:
        print("Opción no válida. Por favor, intenta de nuevo.")
```
